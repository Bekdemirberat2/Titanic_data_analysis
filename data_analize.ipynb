```python
def degiskenler(variable):
    cat=train_data[variable]
    sayi=cat.value_counts()
    plt.figure(figsize=(9,3))
    plt.bar(sayi.index,sayi)
    plt.xticks(sayi.index,sayi.index.values)
    plt.ylabel("frekans")
    plt.title(variable)
    plt.show()
    print(variable,sayi)    
```

```python
category1 = ["Survived","Sex","Pclass","Embarked","SibSp","Parch"]
```

```python
for i in category1:
    degiskenler(i)
```

```python
train_data.info()
```

```python
category2 = ["Cabin","Name","Ticket"]
for i in category2:
    degiskenler(i)
```

```python
def hist(variable):
    plt.figure(figsize=(9,3)) #Figsize ile grafiğin boyutunu ayarlıyoruz.
    plt.hist(train_data[variable],bins=20) #Histogram ile değişkenin dağılımını gösteriyoruz. bins parametresi ile histogramdaki çubuk sayısını belirliyoruz.
    plt.xlabel(variable) 
    plt.ylabel("frekans")
    plt.title(variable)
    plt.show()
    print(variable,train_data[variable].describe()) #$describe() fonksiyonu ile değişkenin istatistiksel özetini gösteriyoruz.
```

```python
sayisal = ["Fare","Age","PassengerId"]
for i in sayisal:
    hist(i)
```
